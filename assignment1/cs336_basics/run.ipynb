```python
from cs336_basics.train import train_bpe
import os
import json
import time

output_dir = "/content/drive/MyDrive/cs336_output"
os.makedirs(output_dir, exist_ok=True)

vocab_path = os.path.join(output_dir, "vocab_owt.json")
merges_path = os.path.join(output_dir, "merges_owt.txt")

intput_path = "/content/drive/MyDrive/cs336_data/owt_train.txt"
vocab_size = 32000
special_tokens = ["<|endoftext|>"]

start_time = time.time()
print("start training on OpenWebText dataset")
vocab,merges = train_bpe(intput_path,vocab_size,special_tokens)
end_time = time.time()
print("training finished")
print(f"Total time used: {end_time - start_time} seconds")
longest_tokens = sorted(vocab.items(), key=lambda x: len(x[1]), reverse=True)
print(f"the longest token is {longest_tokens[0][1]}")
vocab_for_json = {}
for token_id, token_val in vocab.items():
    if isinstance(token_val, bytes):

        vocab_for_json[token_id] = token_val.decode("latin-1")
    else:
        vocab_for_json[token_id] = token_val
voc = max(vocab_for_json.keys())
with open(vocab_path, "w") as f:
    json.dump(vocab_for_json, f, ensure_ascii=False, indent=2)

with open(merges_path, "w", encoding="utf-8") as f:
    for p1, p2 in merges:
        s1 = p1.decode("latin-1") if isinstance(p1, bytes) else str(p1)
        s2 = p2.decode("latin-1") if isinstance(p2, bytes) else str(p2)
        f.write(f"{s1} {s2}\n")

print(f"the document is stored at: {output_dir}")
```
